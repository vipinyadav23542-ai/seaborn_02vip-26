```python
####seaborn is python data visualization library based on matplotlib.
%pip install seaborn
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
import pandas as pd
df = pd.read_csv("mumbai_west_housing.csv")
df.head()
```

```python
print(df.isnull().sum())
print(df.shape)
```

```python
sns.set_style("whitegrid")
sns.set_palette("viridis") ###color
plt.rcParams["figure.figsize"] = (10,6) ##sizze

```

```python
####basic distribution -- price spread out
sns.histplot(data=df,x="Price_Lakhs",kde=True,color="green",bins=30)
plt.title("Distribution - price spread out")
plt.xlabel("price")
plt.ylabel("Number of house")
plt.show()
```

```python
#####most expensive bhk ----compare price

sns.boxplot(data=df,x="BHK",y="Price_Lakhs",palette="Set2",hue="BHK")
plt.title("house price-number of bhk")
plt.xlabel("BHK")
plt.ylabel("price")
plt.show()
```

```python
#####scateer plot----relationship price vs area

sns.scatterplot(data=df,x="Area_sqft",y="Price_Lakhs",hue="Locality",style="BHK",size="BHK",palette="tab10")

plt.title("house price vs area")
plt.xlabel("area_sqft")
plt.ylabel("Price")
plt.legend(bbox_to_anchor=(1.05,1),loc="upper left")
plt.show()
                
```
